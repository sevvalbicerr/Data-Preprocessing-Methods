```python
import numpy as np 
import pandas as pd
```

```python
V1=np.array([1,3,6,7,1,9,15])
V2= np.array([7,5,8,12,2,3,5])
V3=np.array([12,5,6,14,7,2,31])
df=pd.DataFrame(
{"V1": V1,"V2":V2,"V3":V3})
df=df.astype(float)
df
```

 ## Standardization
 
 Standardizasyon veriyi değiştirir fakat özütünü bozmaz, dönüştürme işlemi ise bilgiyi bazen bozar bazen şeklini değiştirir. 

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

## Normalization
#### Değerleri 0-1 aralığına dönüştürmek için kullanılır. 

```python
from sklearn import preprocessing
preprocessing.normalize(df)
```

## Min-Max Scaling

#### İstediğimiz iki aralık arasındaki değerlere dönüşüm istiyorsak kullanılır. 

```python
scaler=preprocessing.MinMaxScaler(feature_range=(10,20))
scaler.fit_transform(df)
```

# Transformation for discrete variables

```python
import seaborn as sbn 
df=sbn.load_dataset('tips')
df.head()
```

### LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbe=LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["new_sex"]=lbe.fit_transform(df["sex"])
df
```

### İstenilen durumu belirterek diğer durumlara 0 değeri atabiliriz. 

```python
df["new_day"]=np.where(df["day"].str.contains("Sun"),1,0)
```

```python
df
```

## Label Encoder ile Çok sınıflı dönüşüm 

```python
lbe.fit_transform(df["day"])
```
